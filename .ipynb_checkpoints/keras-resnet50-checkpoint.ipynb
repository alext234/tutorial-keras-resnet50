```python
import tensorflow as tf
```

```python
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# this step may takes few minutes to download the pretrained model
model = tf.keras.applications.resnet50.ResNet50()
```

```python
# image_file = "data/ibis.jpg"
image_file = "data/cockatoo.jpg"
img = image.load_img(image_file, target_size=(224,224)) # TODO: how can we use a higher resolution?
```

```python
plt.imshow(img)
```

```python
img_array = image.img_to_array(img)
```

```python
# batch of images as input
img_batch = np.expand_dims(img_array, axis=0)
```

```python
print(img_array.shape)
print(img_batch.shape)
```

```python
img_batch_preprocessed = preprocess_input(img_batch)
```

```python
%%time
prediction = model.predict(img_batch_preprocessed)
```

```python
print(decode_predictions(prediction, top=3))
```
